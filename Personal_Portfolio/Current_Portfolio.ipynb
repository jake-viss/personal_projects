```python
# Import required packages for analysis code.
import os
import pandas as pd
import alpaca_trade_api as tradeapi
import requests
from dotenv import load_dotenv
%matplotlib inline
```

```python
# Load environment variables.
load_dotenv()
```

```python
# Assign variables to API keys.
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Check types to make sure keys are correct.
display(type(alpaca_api_key))
type(alpaca_secret_key)
```

```python
# Create Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Set start & end date times for API import.
start_date = pd.Timestamp("2019-09-14",tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-09-14",tz="America/New_York").isoformat()

# Define the tickers we're going to pull.
main_tickers = ["AMD","MSFT","ATVI","AG","SBUX","AZEK","AAPL","ALK","CRON","AMRN","GDXJ","SQ","FXI","OTLY"]

# Set time frame to 1 day.
timeframe = "1D"

# Set number of rows to 1000 to retrieve max rows.
limit_rows = 1000
```

```python
df_portfolio = alpaca.get_barset(
    main_tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=limit_rows).df

df_portfolio
```

```python
# Create a dataframe for just the close prices for each ticker.
closing_prices_df = pd.DataFrame()

# Loop through each ticker in the ticker list, select the close price for each ticker in the df_portfolio, and set it as the ticker in the new dataframe.
for ticker in main_tickers:
    closing_prices_df[ticker] = df_portfolio[ticker]["close"]
    
# For new closing_prices_df, keep only date as index.
closing_prices_df.index = closing_prices_df.index.date

# Preview the df.
closing_prices_df
```

```python
# Create daily returns dataframe for the closing prices.
daily_returns = closing_prices_df.pct_change().fillna(0)

#Preview the new DF.
daily_returns.head()
```

```python
# Plot the daily returns.
daily_returns.plot(figsize=(15,10),title="Current Stocks in Portfolio - Prior 3 Years Daily Returns")
```

```python
# Calculate cumulative returns.
cumulative_returns = (1 + daily_returns).cumprod()

# Plot the cumulative_returns.
cumulative_returns.plot(figsize=(15,10),title="Current Stocks in Portfolio - Prior 3 Years Cumulative Returns")
```

```python
# Get the summary statistics for the daily returns dataframe.
daily_returns.describe()
```

```python
# Grab the heavily weighted stocks from my portfolio and add them to their own df.
main_stock_returns = daily_returns.iloc[:,0:6]
main_stock_returns.head()
```

```python
# Plot the main stocks in the portfolio.
(1+main_stock_returns).cumprod().plot(figsize=(15,10), title="Main Stocks in Porfolio")
```
